```python
def to_num(x):
    if (x[-1]=='M'):
        x=float(x[:-1])*1000000
    elif (x[-1]=="K"):
        x=float(x[:-1])*1000
    return float(x)
ds['numitems_2022']=ds['numitems_2022'].apply(to_num)
```

```python
ds['owner_2022']=ds['owner_2022'].apply(to_num)
```

```python
ds['total_volume_by2022']=ds['total_volume_by2022'].apply(to_num)
```

```python
def to_num_2(x):
    if x=="---":
        return x
    if x=="< 0.001":
        return 0.0005
    if (x[-1]=='M'):
        x=float(x[:-1])*1000000
    elif (x[-1]=="K"):
        x=float(x[:-1])*1000
    return float(x)
ds['min_price_2022']=ds['min_price_2022'].apply(to_num_2)
```

```python
def numitems_clean(x):
    print(x)
    if type(x)!=str:
        return 0
    else:
        if x=="1 item":
            return 1
        if x=="scrapeagain":
            return 0
        if "," in x:
            return int(x[:-6].replace(",",""))
        else:
            return int(x[:-6])
```

```python
ds["numitems_left_2022"]=ds["numitems_left_2022"].apply(numitems_clean)
```

```python
def follower(i):
    print(i,type(i))
    if type(i)==str:
        if i=="1 follower":
            return 1
        if i[-10:]==" followers":
            i=i[:-10]
        i=i.replace(",","")
        if i[-1].lower()=="m":
            return int(float(i[:-1])*1000000)
        if i[-1].lower()=="k":
            return int(float(i[:-1])*1000)
        print("here,",i)
        return int(i)
    else:
        print("wa?")
        return 0
```

```python
ds["ig"]=ds["ig_followers"].apply(follower)
```
